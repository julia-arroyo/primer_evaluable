Lee sobre espirales en http://en.wikipedia.org/wiki/Spiral; luego escribe un programa que dibuje una espiral arquimediana (o de otro tipo). Al igua que antes debes definir preciamente las funciones, qué hace y cómo vas a irlas probando.

```python
import turtle #importo la librería turtle

def espiral(vueltas, t, radio):
    """
    Esta función crea la espiral
    """
    for i in range(36 * vueltas): #el bucle se va a repetir 36 * el número de vueltas deseado
        t.circle((radio * i), 10) #se va a dibujar un arco de radio proporcionado * i (puede valer 0, 1, 2, ..., 3*vueltas) y con un ángulo de 10 grados
        

r = int(input("Introduce el radio de la espiral: ")) #el usuario introduce el radio de la espiral
t = turtle.Turtle() #defino la tortuga
v = int(input("Introduce el número de vueltas: ")) #el usuario introduce el número de vueltas

espiral(v, t, r) #llamo a la funciión
turtle.done()
```
